```python
# Install the xlrd engine if you suspect the file is an older .xls format
# Although the extension is .xlsx, sometimes older software saves .xls content
# with a .xlsx extension.
!pip install xlrd

import pandas as pd
import re

# Assuming the file might be an older .xls format despite the extension
# Try specifying the 'xlrd' engine
try:
    youtube_df = pd.read_excel("/content/youtube_main_comments_with_time11.xlsx", engine='xlrd')
    print("Successfully read file using xlrd engine.")
except Exception as e:
    print(f"Failed to read file with xlrd engine: {e}")
    # If xlrd still fails, try the default engine again
    try:
        print("Trying default engine again...")
        youtube_df = pd.read_excel("/content/youtube_main_comments_with_time11.xlsx")
        print("Successfully read file using default engine.")
    except Exception as e:
        print(f"Failed to read file with default engine as well: {e}")
        print("The file might be corrupted or not a valid Excel file.")


# The rest of your code
def clean_text(text):
    text = str(text)
    text = text.lower()
    text = re.sub(r"http\S+", "", text)  # remove URLs
    text = re.sub(r"@\w+", "", text)     # remove mentions
    text = re.sub(r"[^a-zA-Z\s]", "", text)  # remove special characters
    text = re.sub(r"\s+", " ", text).strip()
    return text

# Check if youtube_df was successfully created before processing
if 'youtube_df' in locals() and isinstance(youtube_df, pd.DataFrame):
    youtube_df['Cleaned_Comment'] = youtube_df['Comment'].apply(clean_text)
    print("Cleaned comments added to DataFrame.")
else:
    print("DataFrame not loaded successfully. Cannot proceed with cleaning.")
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')

sia = SentimentIntensityAnalyzer()

youtube_df['Sentiment_Score'] = youtube_df['Cleaned_Comment'].apply(lambda x: sia.polarity_scores(x)['compound'])

def label_sentiment(score):
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

youtube_df['Sentiment_Label'] = youtube_df['Sentiment_Score'].apply(label_sentiment)
```

```python
youtube_df.head()
```

```python
youtube_df['Date'] = pd.to_datetime(youtube_df['Published Time'])
# Change the column name from an empty string '' to 'Month'
youtube_df['Month'] = youtube_df['Date'].dt.to_period('M').astype(str)
```

```python
monthly_sentiment = youtube_df.groupby('Month').agg({
    'Sentiment_Score': 'mean',
    'Sentiment_Label': lambda x: x.value_counts().to_dict(),
    'Cleaned_Comment': 'count'
}).rename(columns={'Cleaned_Comment': 'Comment_Count'}).reset_index()

monthly_sentiment.head()
```

```python
# @title Sentiment_Score vs Comment_Count

from matplotlib import pyplot as plt
monthly_sentiment.plot(kind='scatter', x='Sentiment_Score', y='Comment_Count', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Sentiment_Score

from matplotlib import pyplot as plt
monthly_sentiment['Sentiment_Score'].plot(kind='hist', bins=20, title='Sentiment_Score')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
